```python
movies_df=movies_df.merge(credits_df,on='title')
```

```python
##removing unwanted columns
movies_df=movies_df[['movie_id','title','genres','keywords','cast','crew','overview']]
```

```python
movies_df.info()
```

```python
movies_df.dropna(inplace=True)
```

```python
##Find duplicate value
movies_df.duplicated().sum()
```

```python
movies_df.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
def convert2(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
        
        return l
```

```python
movies_df['cast']=movies_df['cast'].apply(convert2)
```

```python
movies_df.head()
```

```python
def fetch(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch)
```

```python
movies_df.head()
```

```python
movies_df['overview'][0]
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df.head()
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x: [i.replace(' ', '') if i is not None else i for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [i.replace(' ', '') if i is not None else i for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x: [i.replace(' ', '') if (i is not None and isinstance(i, str)) else i for i in x] if isinstance(x, list) else x)
movies_df['crew'] = movies_df['crew'].apply(lambda x: [i.replace(' ', '') if i is not None else i for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']=movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

```python
movies_df.head()
```

```python
movies=movies_df.drop(columns=['genres','keywords','cast','crew','overview'])
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x) if isinstance(x, (list, tuple)) else x)
```

```python
movies.head()
```

```python
movies['tags'][0]
```

```python
movies['tags']=movies['tags'].apply(lambda x:x.lower() if isinstance(x,str) else x)
```

```python
movies['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
movies['tags'] = movies['tags'].fillna('')
```

```python
cv.fit_transform(movies['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    print('Recommended movies:')
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```

```python
recommend('Avengers: Age of Ultron')
```
